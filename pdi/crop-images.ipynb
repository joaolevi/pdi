```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from utils import *
img_path = '/home/joaolevi/projects/pdi/images/valhala2.webp'
plt.rcParams['image.cmap'] = 'gray'
plt.rcParams["figure.dpi"] = 100
```

```python
def bgr2rgb(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.imread(img_path)
height, width = image.shape[:2]
```

Lets get starting pixel coordinates (top left of cropping rectangle)

```python
start_row, start_col = int(height * .25), int(width * .25)
```

Lets get the ending pixel coordinates (bottom right)

```python
end_row, end_col = int(height * .75), int(height * .75)
```

Simply use indexing to crop out the rectangle we desire

```python
cropped = image[start_row:end_row, start_col:end_col]
```

```python
plt.figure(figsize=(16,8))
plt.subplot(121), plt.imshow(bgr2rgb(image  ))
plt.subplot(122), plt.imshow(bgr2rgb(cropped))
plt.show()
```
