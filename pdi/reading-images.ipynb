```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

path = "images/"
plt.rcParams['image.cmap'] = 'gray'
plt.rcParams['figure.dpi'] = 100
```

```python
def imshow(img):
    cv2.namedWindow('img', cv2.WINDOW_KEEPRATIO)
    cv2.imshow('img', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

def bgr2rgb(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#%% Reading a grayscale image
img_path = '/home/joaolevi/projects/pdi/images/valhala2.webp'
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
plt.imshow(img)
plt.show()
```

```python
img = cv2.imread(img_path)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img = cv2.imread(img_path)
bgr = cv2.split(img)
# b, g, r = cv2.split(img)

plt.figure(figsize = (16,10))
plt.subplot(221); plt.title("original"); plt.imshow(bgr2rgb(img))
plt.subplot(222); plt.title("b"); plt.imshow(bgr[0], cmap='gray')
plt.subplot(223); plt.title("g"); plt.imshow(bgr[1], cmap='gray')
plt.subplot(224); plt.title("r"); plt.imshow(bgr[2], cmap='gray')
plt.show()
```

```python
img = cv2.imread(img_path, cv2.IMREAD_COLOR)

color = ('b', 'g', 'r')

plt.figure(figsize=(16,8))
plt.subplot(121), plt.imshow(bgr2rgb(img))
plt.subplot(122)
for i, col in enumerate(color):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color=col)
    plt.xlim([0, 256])
plt.show()

imshow(img)
```
