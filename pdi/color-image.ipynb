```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../images/baboon.png', cv2.IMREAD_COLOR)
```

```python
b, g, r = cv2.split(img)

r = r.flatten()
g = g.flatten()
b = b.flatten()

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(r, g, b, c=np.vstack((r, g, b)).T / 255, marker='o')

ax.set_xlabel('Red')
ax.set_ylabel('Green')
ax.set_zlabel('Blue')

plt.show()
```

```python
img = cv2.imread('../images/chips.png', cv2.IMREAD_COLOR)
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)

hist_r = cv2.calcHist([img], [0], None, [256], [0, 256])
hist_g = cv2.calcHist([img], [1], None, [256], [0, 256])
hist_b = cv2.calcHist([img], [2], None, [256], [0, 256])
hist_h = cv2.calcHist([hsv], [0], None, [180], [0, 180])
hist_s = cv2.calcHist([hsv], [1], None, [256], [0, 256])
hist_v = cv2.calcHist([hsv], [2], None, [256], [0, 256])

plt.figure(figsize=(12, 8))
plt.subplot(2, 3, 1)
plt.plot(hist_r, color='r')
plt.title('Red')
plt.subplot(2, 3, 2)
plt.plot(hist_g, color='g')
plt.title('Green')
plt.subplot(2, 3, 3)
plt.plot(hist_b, color='b')
plt.title('Blue')
plt.subplot(2, 3, 4)
plt.plot(hist_h, color='m')
plt.title('Hue')
plt.subplot(2, 3, 5)
plt.plot(hist_s, color='c')
plt.title('Saturation')
plt.subplot(2, 3, 6)
plt.plot(hist_v, color='y')
plt.title('Value')
plt.show()
```

```python
lower_red = np.array([0, 100, 100])
upper_red = np.array([10, 255, 255])

lower_green = np.array([40, 100, 100])
upper_green = np.array([80, 255, 255])

lower_blue = np.array([100, 100, 100])
upper_blue = np.array([140, 255, 255])

lower_yellow = np.array([20, 100, 100])
upper_yellow = np.array([30, 255, 255])

lower_orange = np.array([0, 100, 100])
upper_orange = np.array([10, 255, 255])

lower_black = np.array([0, 0, 0])
upper_black = np.array([180, 255, 50])


mask_red = cv2.inRange(hsv, lower_red, upper_red)
mask_green = cv2.inRange(hsv, lower_green, upper_green)
mask_blue = cv2.inRange(hsv, lower_blue, upper_blue)
mask_yellow = cv2.inRange(hsv, lower_yellow, upper_yellow)
mask_orange = cv2.inRange(hsv, lower_orange, upper_orange)
mask_black = cv2.inRange(hsv, lower_black, upper_black)

plt.figure(figsize=(12, 8))
plt.subplot(2, 3, 1)
plt.imshow(mask_red, cmap='gray')
plt.title('Red')
plt.subplot(2, 3, 2)
plt.imshow(mask_green, cmap='gray')
plt.title('Green')
plt.subplot(2, 3, 3)
plt.imshow(mask_blue, cmap='gray')
plt.title('Blue')
plt.subplot(2, 3, 4)
plt.imshow(mask_yellow, cmap='gray')
plt.title('Yellow')
plt.subplot(2, 3, 5)
plt.imshow(mask_orange, cmap='gray')
plt.title('Orange')
plt.subplot(2, 3, 6)
plt.imshow(mask_black, cmap='gray')
plt.title('Black')
plt.show()
```
