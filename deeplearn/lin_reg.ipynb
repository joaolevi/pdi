```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
xlim = (-10, 10)
step = 1

# Geração de dados para a regressão linear
np.random.seed(1)
X = np.arange(xlim[0], xlim[1], step)[:, np.newaxis]  # Criando uma matriz de uma única coluna
y = 2 * X + np.random.randn(len(X), 1)  # Adicionando algum ruído aos dados

# Visualização dos dados
plt.scatter(X, y, color='blue')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Exemplos para Regressão Linear')
plt.grid(True)
plt.show()
```

```python
class LinearRegression:
    def __init__(self, fit_intercept=True):
        self.fit_intercept = fit_intercept
        self.weights = None

    def fit(self, X, y):
        if self.fit_intercept:
            X = np.c_[np.ones(X.shape[0]), X]  # Adiciona uma coluna de 1s para o termo de interceptação

        # Calcula os coeficientes usando a equação normal
        self.weights = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

    def predict(self, X):
        if self.fit_intercept:
            X = np.c_[np.ones(X.shape[0]), X]  # Adiciona uma coluna de 1s para multiplicar pelos weights

        return X.dot(self.weights) # wT.x

    def get_weights(self):
        return self.weights
    
    def mean_squared_error(self, y_true, y_pred):
        return np.mean(np.power(y_true - y_pred, 2))

# Exemplo de uso:
if __name__ == "__main__":
    # Criar e treinar o modelo
    model = LinearRegression()
    model.fit(X, y)

    # Fazer previsões
    y_pred = model.predict(X)

    print(model.mean_squared_error(y, y_pred))
```

```python
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red', linewidth=1)
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Regressão Linear')
plt.grid(True)
plt.show()
```
